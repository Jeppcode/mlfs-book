## <span style="color:#ff5f27;"> üñç Feature View Creation and Retrieving </span>

```python
# Select features for training data.
selected_features = air_quality_fg.select(['pm25', 'date']).join(weather_fg.select_features(), on=['city'])
```

### Feature Views

`Feature Views` are selections of features from different **Feature Groups** that make up the input and output API (or schema) for a model. A **Feature Views** can create **Training Data** and also be used in Inference to retrieve inference data.

The Feature Views allows a schema in form of a query with filters, defining a model target feature/label and additional transformation functions (declarative feature encoding).

In order to create Feature View we can use `FeatureStore.get_or_create_feature_view()` method.

You can specify the following parameters:

- `name` - name of a feature group.

- `version` - version of a feature group.

- `labels`- our target variable.

- `transformation_functions` - declarative feature encoding (not used here)

- `query` - selected features/labels for the model 

```python
feature_view = fs.get_or_create_feature_view(
    name='air_quality_fv',
    description="weather features with air quality as the target",
    version=2,
    labels=['pm25'],
    query=selected_features,
)
```

## <span style="color:#ff5f27;">ü™ù Split the training data into train/test data sets </span>

We use a time-series split here, with training data before this date `start_date_test_data` and test data after this date

```python
start_date_test_data = "2025-05-01"
# Convert string to datetime object
test_start = datetime.strptime(start_date_test_data, "%Y-%m-%d")
```

```python
X_train, X_test, y_train, y_test = feature_view.train_test_split(
    test_start=test_start
)
```

```python
X_train
```

### Added part C for the previous 3 days feature

```python
# Add lagged PM2.5 features (1, 2, 3 days)
# Merge PM2.5 values into X_train and X_test for lag creation
train_full = X_train.copy()
train_full['pm25'] = y_train['pm25']

# Sort by date to ensure correct lagging
train_full = train_full.sort_values('date')

# Create lag features
train_full['pm25_lag_1'] = train_full['pm25'].shift(1)
train_full['pm25_lag_2'] = train_full['pm25'].shift(2)
train_full['pm25_lag_3'] = train_full['pm25'].shift(3)

# Fill NaN values (for first rows without previous data)
train_full.fillna(0, inplace=True)

# Separate features and target again
X_train = train_full.drop(columns=['pm25'])
y_train = train_full[['pm25']]

# Repeat for test set
test_full = X_test.copy()
test_full = test_full.sort_values('date')

# We need previous day's PM2.5, so merge with last 3 values from train set
last_train_pm25 = train_full['pm25'].iloc[-3:].values # grabs the last 3 days of PM2.5 from the training set.
test_full['pm25_lag_1'] = pd.concat([pd.Series([last_train_pm25[-1]]), y_train['pm25'][:len(test_full)-1]]).reset_index(drop=True)
test_full['pm25_lag_2'] = pd.concat([pd.Series(last_train_pm25[-2:]), y_train['pm25'][:len(test_full)-2]]).reset_index(drop=True)
test_full['pm25_lag_3'] = pd.concat([pd.Series(last_train_pm25[-3:]), y_train['pm25'][:len(test_full)-3]]).reset_index(drop=True)
# Each lag column for the test set combines:
# The last PM2.5 values from training to fill the first rows.
# The earlier test PM2.5 values to continue the lag sequence.

test_full.fillna(0, inplace=True)
X_test = test_full
```

```python
X_features = X_train.drop(columns=['date'])
X_test_features = X_test.drop(columns=['date'])
```

## <span style="color:#ff5f27;">üß¨ Modeling</span>

We will train a regression model to predict pm25 using our 4 features (wind_speed, wind_dir, temp, precipitation)

```python
# Creating an instance of the XGBoost Regressor
xgb_regressor = XGBRegressor()

# Fitting the XGBoost Regressor to the training data
xgb_regressor.fit(X_features, y_train)
```

```python
# Predicting target values on the test set
y_pred = xgb_regressor.predict(X_test_features)

# Calculating Mean Squared Error (MSE) using sklearn
mse = mean_squared_error(y_test.iloc[:,0], y_pred)
print("MSE:", mse)

# Calculating R squared using sklearn
r2 = r2_score(y_test.iloc[:,0], y_pred)
print("R squared:", r2)
```

```python
df = y_test
df['predicted_pm25'] = y_pred
```

## Explanation of Performance Change

### Improved Model Accuracy
- **MSE decreased** from 6.936 ‚Üí 4.743 ‚Üí lower error means the predictions are closer to actual PM2.5 values.
- **R¬≤ improved** from -0.651 ‚Üí -0.129 ‚Üí closer to 0, showing the model explains variance in PM2.5 better, though still not perfect.

### Impact of Lagged Features
- `pm25_lag_1`, `pm25_lag_2`, `pm25_lag_3` appear among the top features in the new model.
- These features **capture temporal dependence**: today's PM2.5 is highly correlated with the previous days.
- Adding them reduced reliance on weather features alone, which were the dominant features in the old model.

### Feature Importance Shift
- Weather features like `temperature_2m_mean`, `wind_speed_10m_max`, and `wind_direction_10m_dominant` dropped in importance relative to lagged PM2.5.
- This is expected because past PM2.5 values carry strong predictive signal for near-future PM2.5.
